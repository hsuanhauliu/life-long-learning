# Line Detection - Canny Edge Detector

```python
%matplotlib notebook

import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
ori_img = cv.imread('imgs/test.png')
rgb_img = cv.cvtColor(ori_img, cv.COLOR_BGR2RGB)
gray_img = cv.cvtColor(rgb_img, cv.COLOR_RGB2GRAY)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
ax1.imshow(rgb_img)
ax2.imshow(gray_img, cmap='gray')
fig.show()
```

First, we'll apply Canny edge detector without blurring the image.

```python
canny = cv.Canny(gray_img, 50, 150)
plt.imshow(canny, cmap='gray')
plt.show()
```

As we can see, there's a lot of noise in the image. Now we want to apply Gaussian blurring before applying Canny edge detector.
