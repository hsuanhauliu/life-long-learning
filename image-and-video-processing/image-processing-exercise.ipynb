```python
%matplotlib notebook

import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
filename = "imgs/img1.JPG"
img = cv2.imread(filename)

# OpenCV reads image in as BGR channels so we need to convert it to RGB
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Check the shape of the image
print(img.shape)
```

```python
# invert the image
inverted_img = 255 - img

fig, (ax1, ax2) = plt.subplots(1, 2)
ax1.imshow(img)
ax2.imshow(inverted_img)
fig.show()
```

We can convert the image to gray-scale and look at its color distribution using a histogram.

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
print(img_gray.shape)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
ax1.imshow(img_gray, cmap='gray')
ax2.hist(img_gray.ravel(), 256, [0, 256])
fig.show()
```

Let's quantize the image to different gray level scale.

```python
img_128_gray_level = (img_gray // 2) * 2
img_64_gray_level = (img_gray // 4) * 4
img_8_gray_level = (img_gray // 32) * 32
img_2_gray_level = (img_gray // 128) * 128
```

```python
fig, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4)
ax1.imshow(img_128_gray_level, cmap='gray')
ax2.imshow(img_64_gray_level, cmap='gray')
ax3.imshow(img_8_gray_level, cmap='gray')
ax4.imshow(img_2_gray_level, cmap='gray')
fig.show()
```
