# Image Blurring Algorithm Testing

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

%matplotlib inline
```

## Average Blurring

```python
img = cv2.imread('imgs/dog.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
blur_ave = cv2.blur(img,(5,5))

plt.imshow(img)
plt.show()
plt.imshow(blur_ave)
plt.show()
```

```python
blur_gau = cv2.GaussianBlur(img,(5,5),0)

plt.imshow(img)
plt.show()
plt.imshow(blur_gau)
plt.show()
```

```python
blur_med = cv2.medianBlur(img,5)

plt.imshow(img)
plt.show()
plt.imshow(blur_med)
plt.show()
```

```python
blur_bil = cv2.bilateralFilter(img,9,75,75)

plt.imshow(img)
plt.show()
plt.imshow(blur_bil)
plt.show()
```

```python
plt.imshow(img)
plt.show()

fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax1.imshow(blur_ave)
ax1 = fig.add_subplot(2,2,2)
ax1.imshow(blur_gau)
ax1 = fig.add_subplot(2,2,3)
ax1.imshow(blur_med)
ax1 = fig.add_subplot(2,2,4)
ax1.imshow(blur_bil)
```
